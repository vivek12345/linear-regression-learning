```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 3])
y = np.array([10, 30])

plt.figure()
ax = plt.gca()

table = ax.table(
    cellText=[[1,10], [3, 30]],
    colLabels=['Performance Length', 'Tip Amount'],
    cellLoc='center',
    loc='best'
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(0.8, 1.5)

plt.scatter(x, y)
plt.title("Performance Length vs Tip Amount")
plt.xlabel("Performance Length")
plt.ylabel("Tip Amount")

ax.set_xlim([0, 5])
ax.set_ylim([0, 50])
plt.show()
```

```python
x = np.array([1, 3])
y = np.array([10, 30])

plt.figure()
ax = plt.gca()

table = ax.table(
    cellText=[[1,10], [3, 30]],
    colLabels=['Performance Length', 'Tip Amount'],
    cellLoc='center',
    loc='best'
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(0.8, 1.5)

plt.scatter(x, y)
plt.title("Performance Length vs Tip Amount")
plt.xlabel("Performance Length")
plt.ylabel("Tip Amount")

ax.set_xlim([0, 5])
ax.set_ylim([0, 50])

plt.plot(x, y, "-g")

plt.scatter(2, 20, color="red", s=100)
plt.show()
```

```python
x_new = np.array([1, 3, 2, 1.7, 2.5, 2.7, 1.1, 1.6, 2.1, 3])
y_new = np.array([10, 30, 20, 16, 35, 32, 5, 20, 29, 40])

plt.figure()

plt.scatter(x_new, y_new)
plt.title("Performance Length vs Tip Amount")
plt.xlabel("Performance Length")
plt.ylabel("Tip Amount")

ax = plt.gca()

ax.set_xlim([0, 5])
ax.set_ylim([0, 50])

plt.show()
```

```python
plt.figure()
plt.plot(x, 10*x, "-g")
plt.scatter(x_new, y_new)
plt.title("Performance Length vs Tip Amount")
plt.xlabel("Performance Length")
plt.ylabel("Tip Amount")

ax = plt.gca()

ax.set_xlim([0, 5])
ax.set_ylim([0, 50])

plt.show()
```

```python
plt.figure()
plt.plot(x, 10*x, "--g")
plt.scatter(x_new, y_new)
plt.title("Performance Length vs Tip Amount")
plt.xlabel("Performance Length")
plt.ylabel("Tip Amount")

ax = plt.gca()

ax.set_xlim([0, 5])
ax.set_ylim([0, 50])

m, b = np.polyfit(x_new, y_new, 1)
plt.plot(x_new, m*x_new + b, "-r")

plt.show()
```

```python
plt.figure()
plt.plot(x, 10*x, "-g")
plt.scatter(x_new, y_new)

for i in range(len(x_new)):
  plt.plot([x_new[i], x_new[i]], [y_new[i], 10*x_new[i]], "r--")

errors = y_new - 10*x_new
sse = np.sum(errors**2)

print(f'Sum of Squares of Errors (SSE): {sse}')

plt.title("Performance Length vs Tip Amount")
plt.xlabel("Performance Length")
plt.ylabel("Tip Amount")

ax = plt.gca()

ax.set_xlim([0, 5])
ax.set_ylim([0, 50])

plt.show()
```

```python
plt.figure()
m, b = np.polyfit(x_new, y_new, 1)
plt.plot(x_new, m*x_new+b, "-r")
plt.scatter(x_new, y_new)

for i in range(len(x_new)):
  plt.plot([x_new[i], x_new[i]], [y_new[i], m*x_new[i]+b], "r--")

errors = y_new - (m*x_new+b)
sse = np.sum(errors**2)

print(f'Sum of Squares of Errors (SSE): {sse}')

plt.title("Performance Length vs Tip Amount")
plt.xlabel("Performance Length")
plt.ylabel("Tip Amount")

ax = plt.gca()

ax.set_xlim([0, 5])
ax.set_ylim([0, 50])

plt.show()
```
